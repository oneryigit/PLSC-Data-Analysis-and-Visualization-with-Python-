## barplot and countplot

 Barplot is a general plot that allows you to aggregate the categorical data(default: mean)

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np

sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#countplot
#Same with above but it is about the count of occurance

sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

```python
#spine control
sns.countplot(x='sex',data=tips)
sns.despine(left=True, right=False)
#defult is right True. You can pass arguments.
```

```python
#figsize
plt.figure(figsize=(10,3))
sns.countplot(x='sex',data=tips)
```

## stripplot and swarmplot
The `stripplot` will draw a scatterplot where one variable is categorical.

The `swarmplot` is similar to `stripplot()`, but the points are adjusted so that they do not overlap.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1")
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

# Grid and FacetGrid

Grids allow you to map plot types to rows and columns of a grid.

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
sns.PairGrid(iris)
# think of this as a plt.subplots()
# Now, we will map on these axes.
```

```python
# So we could save the grid like we do fig, ax= plt.subplots()

# I save the empty grid as g
g=sns.PairGrid(iris)

#you spesify what you want as plot type
#relations between each variables

g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#pairplot we saw above is one example of grid

sns.pairplot(iris,hue='species',palette='Set1')
```

### Facedgrid

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')

g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

# Regression plots

Seaborn is has builtin capability to do simple regression and provide you the graph.

```python
# lmplot()

sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
# You can use grid as well. Spesify the column or/and row

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
#both rows and columns are provided

sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.5)
```
